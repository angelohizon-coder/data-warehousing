```python
customer_profile2 = pd.read_csv("Mall_Customers - cust_profile2.csv")
```

```python
customer_profile2.to_sql("customer_profile2", engine, index=False, if_exists="replace")
```

```python
query = """
    SELECT * FROM customers INNER JOIN customer_profile2 ON customers.customer_id = customer_profile2.cust_id
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT * FROM customers LEFT JOIN customer_profile2 ON customers.customer_id = customer_profile2.cust_id
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT * FROM customers RIGHT JOIN customer_profile2 ON customers.customer_id = customer_profile2.cust_id
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT * FROM customers 
    UNION
    SELECT * FROM customer_profile2
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# assumed to have no intersection in MySQL which is why intersect is similar to inner join 
query = """
    SELECT * FROM customers INNER JOIN customer_profile2 WHERE customer_id = cust_id
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT customer_id, age, education, gender FROM customer_profile2 INNER JOIN customers ON customer_id = cust_id WHERE
    age BETWEEN 30 AND 40 LIMIT 5
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT customer_id, age, education, gender FROM customer_profile2 INNER JOIN customers ON customer_id = cust_id WHERE
    age < 40 AND gender LIKE 'Female' AND education LIKE 'college'
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT customer_id, age, SUM(annual_income) AS sum_annual_income FROM customer_profile2 
    INNER JOIN customers ON customer_id = cust_id
    GROUP BY customer_id, age
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT customer_id, siblings, SUM(siblings) AS sum_siblings 
    FROM customer_profile2 INNER JOIN customers ON customer_id = cust_id
    GROUP BY customer_id, siblings
    ORDER BY SUM(siblings)
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# NO CUBE IN MYSQL
query = """
    SELECT 
        customer_id, age, annual_income, AVG(annual_income) 
    FROM 
        customer_profile2 
    INNER JOIN 
        customers 
    ON 
        customer_id = cust_id 
    GROUP BY 
        CUBE(cust_id, age, annual_income)
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# Rollup
query = """
    SELECT 
        customer_id, age, AVG(annual_income) 
    FROM 
        customer_profile2 
    INNER JOIN 
        customers 
    ON 
        customer_id = cust_id 
    GROUP BY 
        customer_id, age, annual_income
    WITH ROLLUP
"""

df = pd.read_sql_query(query, conn)
df
```

```python
# Rollup
query = """
    SELECT 
        cust_id, customer_profile2.spending_score, customer_profile2.annual_income, AVG(customer_profile2.annual_income), education
    FROM 
        customer_profile2 
    INNER JOIN 
        customers 
    ON 
        customer_id = cust_id 
    GROUP BY 
        cust_id, customer_profile2.spending_score, customer_profile2.annual_income
    WITH ROLLUP
"""

df = pd.read_sql_query(query, conn)
df
```

```python
query = """
    SELECT * FROM customers FULL JOIN customer_profile2 WHERE customer_id = cust_id
"""

df = pd.read_sql_query(query, conn)
df
```
