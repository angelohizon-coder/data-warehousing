```python
# Hizon
query = ('''
    SELECT 
    cust_id AS id, siblings, SUM(siblings) AS sum_siblings
    FROM customer_profile2
    GROUP BY cust_id, siblings 
    ORDER BY SUM(siblings)
''')
pd.read_sql(query, conn)
```

```python
# Hizon
query = ('''
    SELECT customer_id, age, AVG(spending_score)
    FROM customers
    GROUP BY customer_id, age
    UNION 
    SELECT customer_id, age, AVG(spending_score) 
    FROM customers
    GROUP BY customer_id, age
''')
pd.read_sql(query, conn)
```

```python
# Hizon
query = ('''
    SELECT customer_id, age, AVG(spending_score) AS avg_spending_score
    FROM customers
    GROUP BY customer_id, age
    UNION 
    SELECT customer_profile2.cust_id, MAX(customers.age), 
        AVG(customer_profile2.spending_score)
    FROM customer_profile2
    INNER JOIN customers
    ON customer_profile2.cust_id = customers.customer_id
    GROUP BY customer_profile2.cust_id
    ORDER BY customer_id
''')
c.execute(query)
c.fetchall()
```

```python
query6 = ('''
SELECT
    cust_id, age, 
    customer_profile2.annual_income, 
    AVG(customer_profile2.annual_income) as average_income
FROM 
    customers INNER JOIN customer_profile2
ON
    customers.customer_id = customer_profile2.cust_id
GROUP BY
    ROLLUP(cust_id,age,customer_profile2.annual_income)
ORDER BY
    cust_id
''')
pd.read_sql(query6, conn)
```

```python
query7 = ("""
    SELECT
        cust_id,
        spending_score,
        annual_income,
        AVG(annual_income) AS average_income    
    FROM
        customer_profile2
    GROUP BY
        education,
        CUBE(cust_id, spending_score, annual_income)
    ORDER BY education
""")
pd.read_sql(query7, conn)
```
