```python
query = ('''
    SELECT 
    id, gender, gwa, avg(gwa)
    FROM students_3csc 
    GROUP BY id, gender, gwa
''')
c.execute(query)
c.fetchall()
```

```python
query = ('''
    SELECT 
    gender, AVG(gwa)
    FROM students_3csc 
    GROUP BY gender
''')
c.execute(query)
c.fetchall()
```

```python
query = ('''
    SELECT 
    age, AVG(gwa)
    FROM students_3csc 
    GROUP BY age
    ORDER BY avg(gwa) desc
''')
c.execute(query)
c.fetchall()
```

```python
query = ('''
    SELECT 
    strand, ustet_score, COUNT(shs_gwa)
    FROM students_3csc_grades 
    GROUP BY CUBE(strand, ustet_score)
''')
pd.read_sql(query, conn)
```

```python
query = ('''
    SELECT 
    strand, ustet_score, jhs_gwa, SUM(ustet_score)
    FROM students_3csc_grades 
    GROUP BY strand, CUBE(ustet_score, jhs_gwa) ORDER BY strand
''')
pd.read_sql(query, conn)
```

```python
#NaN is for numerical and None is for categorica
query = ('''
    SELECT 
    strand, ustet_score, jhs_gwa, SUM(ustet_score), COUNT(strand)
    FROM students_3csc_grades 
    GROUP BY strand, jhs_gwa, CUBE(ustet_score) ORDER BY strand
''')
pd.read_sql(query, conn)
```

```python
#NaN is for numerical and None is for categorica
query = ('''
    SELECT 
    gender, gwa, age, AVG(gwa)
    FROM students_3csc
    GROUP BY ROLLUP(gender, gwa, age)
    ORDER BY gender
''')
pd.read_sql(query, conn)
```

```python
#NaN is for numerical and None is for categorical
query = ('''
    SELECT 
    gender, gwa, age, AVG(gwa)
    FROM students_3csc
    GROUP BY gender, age, ROLLUP(gwa)
    ORDER BY gender
''')
pd.read_sql(query, conn)
```

```python
#NaN is for numerical and None is for categorical
query = ('''
    SELECT 
    gender, gwa, age, AVG(gwa)
    FROM students_3csc
    WHERE age > 20 AND gwa < 2
    GROUP BY age, ROLLUP(gwa, gender)
    ORDER BY gender
''')
pd.read_sql(query, conn)
```
