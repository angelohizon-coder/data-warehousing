```python
from petl import look, fromdb, fromjson,  unpackdict, fromdicts, cut, rename, tocsv
from pandas import DataFrame
import numpy as np
import pandas as pd
import requests, urllib, json
```

```python
url = urllib.request.urlopen('https://jsonplaceholder.typicode.com/users')
users = json.load(url)
users
```

```python
#extraction
user_info = fromdicts(users)
user_info
```

```python
#unpacking
user_info = unpackdict(user_info, 'address')
user_info = unpackdict(user_info, 'geo')
```

```python
user_info = cut(user_info, 'id', 'name', 'username', 'email', 'street', 
                'suite', 'city', 'zipcode', 'lat', 'lng', 'phone', 'website')
user_info
```

```python
#renames the header
user_info = rename(user_info, {'id': 'employee_id', 'name': 'complete_name'})
user_info
```

```python
#extraction
user_info_v2 = fromdicts(users)
user_info_v2
```

```python
#unpacking
user_info_v2 = unpackdict(user_info_v2, 'address')
user_info_v2 = unpackdict(user_info_v2, 'geo')
user_info_v2 = unpackdict(user_info_v2, 'company')
```

```python
user_info_v2 = cut(user_info_v2, 'id', 'name', 'username', 'email', 'street', 
                'suite', 'city', 'zipcode', 'lat', 'lng', 'phone', 'website',
                 'name', 'catchPhrase', 'bs')
user_info_v2
```

```python
user_info_v2 = rename(user_info_v2, {'id': 'employee_id', 'name': 'complete_name', 'catchPhrase': 'catch phrase', 'bs': 'business solution'})
user_info_v2
```

```python
tocsv(user_info_v2, 'user.csv')
```
