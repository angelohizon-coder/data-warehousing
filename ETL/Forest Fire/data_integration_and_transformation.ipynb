```python
#Dimension reduction (principal component analysis)
from sklearn.decomposition import PCA
```

```python
#Compresses x values from 2:13 to 5 
pca = PCA(n_components = 5)
pca_ff = pca.fit_transform(x)
new_ff_x = pd.DataFrame(data = pca_ff, columns = ['x1', 'x2', 'x3', 'x4', 'x5'])
new_ff_x.head()
```

```python
#Compresses x values from 2:13 to 6 
pca = PCA(n_components = 6)
pca_ff = pca.fit_transform(x)
new_ff_x = pd.DataFrame(data = pca_ff, columns = ['x1', 'x2', 'x3', 'x4', 'x5', 'x6'])
new_ff_x.head()
```

```python
#Steps in performing the PCA
#1. Standardization 
ff_std_x = StandardScaler().fit_transform(x)
ff_std_x
```

```python
#2 Compute the covariance matrix
mean_ff = np.mean(ff_std_x, axis = 0)
```

```python
mean_ff
```

```python
# number of n (ff_std_x.shape[0]-1)
cov_ff = (ff_std_x - mean_ff).T.dot((ff_std_x - mean_ff))/(ff_std_x.shape[0]-1)
cov_ff
```

```python
#2.1 Alternative covariance
#T means transpose
cov_ff2 = np.cov(ff_std_x.T)
cov_ff2
```

```python
#3 Compute the eigenvalues and eigenvectors of the covariance metrix to 
    #identify the principal components eig_vals
eig_vals, eig_vecs = np.linalg.eig(cov_ff)
eig_vals
```

```python
eig_vecs
```

```python
#4 Create a featiure vector to decide which principal components to retain
#5 recast the data along the principal component
pca2 = PCA(n_components = 5)
pca_ff2 = pca2.fit_transform(ff_std_x)
new_ff_x2 = pd.DataFrame(data = pca_ff2, columns = ['x1', 'x2', 'x3', 'x4', 'x5'])
new_ff_x2.head()
```
